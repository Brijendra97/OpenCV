```python
import cv2
```

```python
cap = cv2.VideoCapture(0) 
# 0 for inbuilt camera
# otherwise the file name will be the argument

while(True):
    ret,frame = cap.read() # ret is return (True or False)
                           # cap.read() it read frame from webcam
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) == ord('q'): # 1 will read continuous frames
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# Read as gray 
cap = cv2.VideoCapture(0)
print(cap.isOpened()) # Gives True if Correct
while(True):
    ret,frame = cap.read()
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    # here cvtColor used for color 
    # frame is source
    cv2.imshow('MyCam',gray)
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
