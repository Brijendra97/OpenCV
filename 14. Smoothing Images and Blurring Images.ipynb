```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('Images/opencv-logo.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
titles = ['image']
images = [img]
```

```python
# filters
# Homogeneous filter, Gaussian filter, Median filter, Bilateral filter
```

```python
# Homogeneous filter :
# simple filter
# each output pixel is the mean of its kernel neighbors
```

```python
kernel = np.ones((5,5),np.float32)/25 # 25 = 5 * 5
dst = cv2.filter2D(img,-1,kernel)

titles = ['image','2D Convolution']
images = [img,dst]

for i in range(2):
    plt.figure(figsize=(18,12))
    plt.subplot(2,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# as in one-dimension signals, images also can be filtered with
# with various low-pass filters(LPF), high-pass filters(HPF) etc
```

```python
# LPF helps in removing noises, blurring the images
# HPF filters helps in finding edges in the images
```

```python
# 1. Averaging
kernel = np.ones((5,5),np.float32)/25 # 25 = 5 * 5
dst = cv2.filter2D(img,-1,kernel)
blur = cv2.blur(img,(5,5))

titles = ['image','2D Convolution','blur']
images = [img,dst,blur]

for i in range(3):
    plt.figure(figsize=(18,12))
    plt.subplot(3,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# 2. Gaussian filter
# use different wight kernel in both x and y direction
# 1. Averaging
kernel = np.ones((5,5),np.float32)/25 # 25 = 5 * 5
dst = cv2.filter2D(img,-1,kernel)
blur = cv2.blur(img,(5,5))
gblur = cv2.GaussianBlur(img,(5,5),0)

titles = ['image','2D Convolution','blur','gblur']
images = [img,dst,blur,gblur]

for i in range(4):
    plt.figure(figsize=(18,12))
    plt.subplot(4,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# 3. Median filter
# it replace each pixel's value with the median of its
# neighboring pixels. This method is great when dealing
# with 'salt and pepper noise'
```

```python
img = cv2.imread('Images/water.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernel = np.ones((5,5),np.float32)/25 # 25 = 5 * 5
dst = cv2.filter2D(img,-1,kernel)
blur = cv2.blur(img,(5,5))
gblur = cv2.GaussianBlur(img,(5,5),0)
median = cv2.medianBlur(img,5) # size must be odd number, not 1

titles = ['image','2D Convolution','blur','gblur','median']
images = [img,dst,blur,gblur,median]

for i in range(5):
    plt.figure(figsize=(18,12))
    plt.subplot(5,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# 4. Bilateral filter ( preserve border)
img = cv2.imread('Images/lena.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernel = np.ones((5,5),np.float32)/25 # 25 = 5 * 5
dst = cv2.filter2D(img,-1,kernel)
blur = cv2.blur(img,(5,5))
gblur = cv2.GaussianBlur(img,(5,5),0)
median = cv2.medianBlur(img,5) # size must be odd number, not 1
bilateralFilter = cv2.bilateralFilter(img,9,75,75)

titles = ['image','2D Convolution','blur','gblur','median','bilateralFilter']
images = [img,dst,blur,gblur,median,bilateralFilter]

for i in range(6):
    plt.figure(figsize=(18,12))
    plt.subplot(6,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
