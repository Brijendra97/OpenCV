```python
# Image gradient is a directional change in the 
# intensity or color in an image.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('Images/messi5.jpg',0)
titles = ['image']
images = [img]
for i in range(1):
    plt.subplot(1,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# we have 3 gradients methods
```

```python
# 1. Laplacian
img = cv2.imread('Images/sudoku.png',0)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize=3)
lap = np.uint8(np.absolute(lap)) # convert to unsigned 

titles = ['image','lap']
images = [img,lap]
for i in range(2):
    plt.figure(figsize=(12,8))
    plt.subplot(2,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# 2. SobelX 3. SobelY
img = cv2.imread('Images/sudoku.png',0)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize=3)
lap = np.uint8(np.absolute(lap)) # convert to unsigned 
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0) # intensity change in x direction
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1) # intensity change in y direction

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

titles = ['image','lap','sobelX','sobelY']
images = [img,lap,sobelX,sobelY]
for i in range(4):
    plt.figure(figsize=(20,12))
    plt.subplot(4,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```

```python
# combine sobelx and sobely
# 2. SobelX 3. SobelY
img = cv2.imread('Images/sudoku.png',0)


sobelX = cv2.Sobel(img,cv2.CV_64F,1,0) # intensity change in x direction
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1) # intensity change in y direction

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX,sobelY)

titles = ['image','sobelX','sobelY','sobelCombined']
images = [img,sobelX,sobelY,sobelCombined]
for i in range(4):
    plt.figure(figsize=(20,12))
    plt.subplot(4,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
